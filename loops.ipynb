```python
import random
while True:
    
    ran_int = random.randint(1,50)
    guessing_chance = 5
    while guessing_chance>0 :
        guess= int(input('enter a number : '))
        if guess==ran_int :
            print('CONGARATS! YOU HAVE WON!')
            break
        elif guess>ran_int :
            print('You have entered a larger number :')
        else :
            print('You have entered a smaller number :')
    
        if guessing_chance == 1 :
            print('YOU HAVE LOST!\n The number was :',ran_int)
        guessing_chance -= 1
    x=int(input('WOULD YOU LIKE TO PLAY AGAIN\n Enter 1 for yes\n Enter 2 to exit \n :'))
    if x==1 :
        print('GOOD ! LET\'S CONTINUE :')
    elif x==2 :
        print('EXITING!')
        break
    else :
        print('invalid input ! exiting')
        break
    
    
```

```python
#printing double pyramid pattern using while loop
i=1

while i <= 5 :
    j = 1
    k = 5
    while k >= i :
        print('',end = ' ')
        k -= 1
    
    while j <= i:
        print('*',end = ' ')
        j +=1
    print('\r')
    i+=1
i=4
while i > 0 :
    k = 4
    print('',end=' ')
    while k >= i :
        print('',end = ' ')
        k -=1
    j = i
    while j > 0 :
        print('*', end = ' ')
        j -= 1
    print('\r')
    i -=1
        
        
```

```python
#printing table of 24, 50, 29 using loop
i = 1
while i <= 10 :
    print(24, ' * ', i, ' = ', 24*i, end =' ')
    print('| ',50, ' * ', i, ' = ', 50*i, end = ' ')
    print('| ', 29, ' * ', i, ' = ', 29*i, end = ' ')
    print('\n')
    i += 1
```

```python
# program for factorial
user_in = int(input('Enter a number to calculate factorial : '))
fac=1
while user_in > 1 :
    fac *= user_in
    user_in -=1

print('factorial of number is : ', fac)
```

```python
#program to print fibonacchi series
user = int(input('Enter maximum value to print fibonacchi series : '))
print('0 1 1 2 ', end = ' ')
pre_term = 1
curr_term =2
while  curr_term<user :
    x = curr_term + pre_term
    #print('\ncurrent x : ', x)
    if x > user :
        break
    else :
        print(x, end = ' ')
    pre_term = curr_term
    curr_term = x
    

    
```

```python
#program to print whether the entered number is prime or not
a = int(input('Enter number greqter than 1 to check : '))
i = 1
_sum = 0

while i>(a-1) :
    if a%i ==0 :
        
        _sum += 1
    if _sum == 2 :
        break
    i += 1
    
if a == 1 :
    print('1 is neither prime nor composite !')
elif _sum == 1 : 
    print('The number is prime !')
else :
    print('number is composite !')
    
```
